```python
import pandas as pd
```

```python
campus_size = pd.read_csv('campus_size.csv')
```

```python
campus_size = campus_size[['College or University', 'Acres']]
```

```python
standard_names = ['Princeton University',
 'Massachusetts Institute of Technology',
 'Harvard University',
 'Stanford University',
 'Yale University',
 'California Institute of Technology',
 'Duke University',
 'Johns Hopkins University',
 'Northwestern University',
 'University of Pennsylvania',
 'Cornell University',
 'University of Chicago',
 'Brown University',
 'Columbia University',
 'Dartmouth College',
 'University of California, Los Angeles',
 'University of California, Berkeley',
 'Rice University',
 'University of Notre Dame',
 'Vanderbilt University',
 'Carnegie Mellon University',
 'University of Michigan, Ann Arbor',
 'Washington University in St Louis',
 'Emory University',
 'Georgetown University',
 'University of Virginia-Main Campus',
 'University of North Carolina at Chapel Hill',
 'University of Southern California',
 'University of California, San Diego',
 'New York University',
 'University of Florida',
 'The University of Texas at Austin',
 'Georgia Institute of Technology-Main Campus',
 'University of California, Davis',
 'University of California, Irvine',
 'University of Illinois Urbana-Champaign',
 'Boston College',
 'Tufts University',
 'University of California, Santa Barbara',
 'University of Wisconsin, Madison',
 'Boston University',
 'Ohio State University-Main Campus',
 'Rutgers University, New Brunswick',
 'University of Maryland, College Park',
 'University of Rochester',
 'Lehigh University',
 'Purdue University-Main Campus',
 'University of Georgia',
 'University of Washington-Seattle Campus',
 'Wake Forest University']
```

```python
standard_names = [standard_names[i].replace(", ", "-") for i in range(0, len(standard_names))]
```

```python
campus_size = campus_size[campus_size['College or University'].isin(standard_names)]
campus_size['Acres'] = campus_size['Acres'].str.replace(',', '').astype(float)
columbia_entry = {'College or University': 'Columbia University', 'Acres': 36.0}
campus_size = campus_size.append(columbia_entry, ignore_index=True)
```

```python
campus_size.to_csv('campus_data.csv', index=False)
```
