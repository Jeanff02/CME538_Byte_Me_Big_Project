```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pylab as plt
import seaborn as sns
import plotly.express as px
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import AgglomerativeClustering
import tkinter as tk
from tkinter import messagebox
```

```python
# Load the dataset
uni_data = pd.read_csv('merged_datasets.csv')
uni_data.drop(columns=['In State Tuition', 'Graduation Rate 4-years', 'Public or Private', 'ar score', 'isr score', 'ifr score', 'fsr score','Facilities', 'University ID', 'Admission Rate'], inplace=True)

# Define user preferences mapping
preference_mapping = {
    "Very Important": 1,
    "Moderately Important": 0.75,
    "Slightly Important": 0.5,
    "Not Important": None
}

# Define a list of preferences, corresponding columns, and custom logic for "Very Important"
preferences = [
    ("Out of State Tuition", "How important is tuition to you?", lambda df: df['Out of State Tuition'].min(), 0.25),
    ("Number of Students", "How important is it that the school has a small student size?", lambda df: df['Number of Students'].min(), 0.25),
    ("Diversity Score", "How important is the school's ethnic diversity?", lambda df: df['Diversity Score'].max(), 0.75),
    ("Earnings After 10 Years", "How important are the earnings you expect to make ten years after graduation?", lambda df: df['Earnings After 10 Years'].max(), 0.75),
    ("Safety", "How important is school safety to you?", lambda df: df['Safety'].max(), 0.75),
    ("Social", "How important is social life to you?", lambda df: df['Social'].max(), 0.75),
    ("Research Score", "How important is the university's research output to you?", lambda df: df['Research Score'].max(), 0.75),
    ("Average Sentiment", "How important are actual student opinions on this school?", lambda df: df['Average Sentiment'].max(), 0.75),
    ("Retention Rate", "How important is the retention rate to you?", lambda df: df['Retention Rate'].max(), 0.75),
    ("er score", "How important is the school's employer reputation?", lambda df: df["er score"].max(), 0.75),
    ("Acres", "How important is campus size to you?", lambda df: df['Acres'].max(), 0.75)
]

# List to store user input
user_input = {}
feature_weights = {}

# Function to handle button clicks
def on_button_click(preference, column, logic, quantile, answer):
    if answer == "Very Important":
        user_input[column] = logic(uni_data) # Use specific logic for "Very Important"
        feature_weights[column] = preference_mapping[answer]
    elif answer == "Moderately Important":
        user_input[column] = uni_data[column].quantile(quantile)  # Custom quantile logic
        feature_weights[column] = preference_mapping[answer]
    elif answer == "Slightly Important":
        user_input[column] = uni_data[column].mean()
        feature_weights[column] = preference_mapping[answer]
    elif answer == "Not Important":
        uni_data.drop(columns=[column], inplace=True)
    
    # Move to next preference or show the results
    show_next_preference()

def show_next_preference():
    # Clear the previous widgets
    for widget in root.winfo_children():
        widget.destroy()

    # If there are more preferences to show
    if preferences:
        # Get the next preference to show
        column, question, logic, quantile = preferences.pop(0)
        
        # Display the question
        label = tk.Label(root, text=question)
        label.pack()

        # Create buttons for choices
        for choice in ["Very Important", "Moderately Important", "Slightly Important", "Not Important"]:
            button = tk.Button(root, text=choice, command=lambda choice=choice: on_button_click(preference_mapping[choice], column, logic, quantile, choice))
            button.pack()

    # If no more preferences, display the final result
    else:
        final_label = tk.Label(root, text=f"Your preferences have been recorded. User Input: {user_input}")
        final_label.pack()

# Set up the main window
root = tk.Tk()
root.title("University Preferences")

# Start the process
show_next_preference()

# Run the Tkinter event loop
root.mainloop()
```

# Heat Map of Data Columns

```python
# Creates a heatmap for assessing the correlation of each of the features selected
fig, axs = plt.subplots( figsize=(20, 10))
uni_data_corr = uni_data.select_dtypes(include='number')
sns.heatmap(uni_data_corr.corr().abs(), annot=True, square=True, annot_kws={'size': 14})   
axs.tick_params(axis='x', labelsize=14)
axs.tick_params(axis='y', labelsize=14) 
axs.set_title('Correlations', size=15)
plt.show()
```

# K-means Clustering

```python
features = uni_data.select_dtypes(include=['number']).columns.tolist()

# Initialize the scaler
scaler = StandardScaler()
uni_scaler = pd.DataFrame(scaler.fit_transform(uni_data[features]), columns=features)

for column, factor in feature_weights.items():
    if column in uni_scaler.columns:
        uni_scaler[column] = uni_scaler[column] * factor

# Elbow Method to find the optimal number of clusters
inertia = []  # List to store the inertia values for each number of clusters

# Loop through different numbers of clusters (from 1 to 10 in this case)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(uni_scaler)
    inertia.append(kmeans.inertia_)

# Plot the inertia values to visualize the "elbow"
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Within-cluster Sum of Squares)')
plt.show()

# Fit k-means clustering directly on the standardized features
kmeans = KMeans(n_clusters=5, random_state=42)
uni_scaler["Cluster"] = kmeans.fit_predict(uni_scaler)
```

# Feature Importance from a Random Forest

```python
# Fit a random forest model on the cluster labels
rf = RandomForestClassifier(random_state=42)
rf.fit(uni_scaler[features], uni_scaler["Cluster"])

# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame to display the feature importances
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importances
}).sort_values(by="Importance", ascending=False)

print(importance_df)
```

# Feature Importance from Centroid

Features with higher absolute values (positive or negative) in the centroid will have a larger impact on how the clusters are formed. If a feature has a larger difference between its values across clusters, it is more influential.

```python
# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Create a DataFrame to easily view feature importance for each cluster
centroid_df = pd.DataFrame(centroids, columns=features)

# Display the centroids
print(centroid_df)
```

# Clustering 3D Visualization

```python
# Define the features used for clustering
features_and_cluster = importance_df['Feature'].head(3).tolist()
features_and_cluster.append("Cluster")

# Create a DataFrame for plotting
plot_data = uni_scaler[features_and_cluster]

plot_data["Cluster"] = plot_data["Cluster"].astype(str)  # Convert clusters to strings for color mapping


# Create the 3D scatter plot
fig = px.scatter_3d(plot_data,
    x=features_and_cluster[0],  # "Earnings After 10 Years"
    y=features_and_cluster[1],  # "Diversity Score"
    z=features_and_cluster[2],  # "Safety"
    color="Cluster",  # Use cluster labels for color
    title="3D Clustering Visualization",
    labels={
        features_and_cluster[0]: features_and_cluster[0],
        features_and_cluster[1]: features_and_cluster[1],
        features_and_cluster[2]: features_and_cluster[2],
        "Cluster": "Cluster Label"
    })

# Update layout for better visualization
fig.update_traces(marker=dict(size=6))
fig.update_layout(
    scene=dict(
        xaxis_title=features_and_cluster[0],
        yaxis_title=features_and_cluster[1],
        zaxis_title=features_and_cluster[2]
    )
)

# Show the plot
fig.show()
```
