```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Set up Selenium WebDriver with options
options = webdriver.ChromeOptions()
# options.add_argument('--headless')  # Uncomment to run in headless mode if you want

driver = webdriver.Chrome(options=options)
url = "https://www.ratemyprofessors.com/school/1072"
driver.get(url)

# Wait for the page to load
wait = WebDriverWait(driver, 20)  # Increased timeout to 20 seconds

def close_pop_up_ad():
    """Handles closing pop-up ads if they appear at any point."""
    try:
        # Try switching to the iframe if it's blocking the interaction
        iframe = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "iframe[name='IL_SR_FRAME2']")))
        driver.switch_to.frame(iframe)
        print("Switched to iframe with ad.")
        
        # Now try to close the ad by clicking the close button (assuming the close button exists)
        ad_close_button = wait.until(EC.element_to_be_clickable((By.ID, "bx-close-inside-1177612")))
        ad_close_button.click()
        print("Ad pop-up closed.")
        
        # Switch back to the main content
        driver.switch_to.default_content()
        print("Switched back to main content.")
    except:
        print("No ad iframe detected or could not close it.")

def close_cookie_popup():
    """Handles closing the cookie pop-up if it appears."""
    try:
        cookies_button = wait.until(
            EC.element_to_be_clickable(
                (By.XPATH, "//button[contains(@class, 'CCPAModal__StyledCloseButton-sc-10x9kq-2')]")
            )
        )
        cookies_button.click()
        print("Cookies pop-up dismissed.")
    except:
        print("No cookies pop-up detected.")

def click_show_more():
    """Handles clicking the 'Show More' button and ensuring new content loads."""
    try:
        # Wait for the "Show More" button to be visible and clickable
        show_more_button = wait.until(
            EC.visibility_of_element_located(
                (By.XPATH, "//button[contains(@class, 'Buttons__Button-sc-19xdot-1') and contains(@class, 'PaginationButton__StyledPaginationButton-txi1dr-1')]")
            )
        )

        # Check if the button is enabled (in case it's temporarily disabled)
        if show_more_button.is_enabled():
            print("Clicking 'Show More' button.")
            show_more_button.click()
            time.sleep(3)  # Wait for reviews to load after clicking "Show More"
            return True
        else:
            print("The 'Show More' button is not clickable.")
            return False
    except Exception as e:
        print(f"Error while trying to click 'Show More': {e}")
        return False

try:
    # Handle cookie pop-up if it appears at the beginning
    close_cookie_popup()

    # Handle the pop-up ad with the specified ID at the beginning
    close_pop_up_ad()

    previous_reviews_count = 0  # Initialize counter to track progress
    reviews_scraped = 0  # Keep track of how many reviews we've scraped
    
    try:
        while True:
            # Handle pop-up ad before each "Show More" click, in case an ad pops up
            close_pop_up_ad()

            # Attempt to click 'Show More' button
            if not click_show_more():
                print("No more 'Show More' buttons found, ending scraping.")
                break

            # Simulate scrolling down to ensure new content is triggered
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            # Get the current number of reviews after loading more
            page_source = driver.page_source
            soup = BeautifulSoup(page_source, "html.parser")
            reviews_container = soup.find(id="schoolRatingsList")
            if reviews_container:
                reviews = reviews_container.find_all("li")
                reviews_scraped = len(reviews)
                print(f"Scraped {reviews_scraped} reviews so far.")
                
            # Break if no new reviews are loaded (if the count doesn't change)
            if reviews_scraped == previous_reviews_count:
                print("No new reviews loaded.")
                break
            
            # Update the counter
            previous_reviews_count = reviews_scraped
                
    finally:
        # After scraping, print out all reviews
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")

        reviews_container = soup.find(id="schoolRatingsList")
        if reviews_container:
            reviews = reviews_container.find_all("li")
            print(f"Total reviews scraped: {len(reviews)}")
            for review in reviews:
                print(review.text.strip())
        else:
            print("No reviews found.")
    
        driver.quit()

except Exception as e:
    print(f"An error occurred: {e}")
    driver.quit()
```
