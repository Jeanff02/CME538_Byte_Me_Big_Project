```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pylab as plt
import seaborn as sns
import plotly.express as px
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import AgglomerativeClustering
import tkinter as tk
from tkinter import messagebox
```

```python
uni_data = pd.read_csv('merged_datasets.csv')
uni_data.drop(columns = ['University ID'], inplace=True)
uni_data.head()
```

```python
# Creates a heatmap to show correlation
fig, axs = plt.subplots(figsize=(8, 8))  
uni_data_corr = uni_data.select_dtypes(include='number')
sns.heatmap(uni_data_corr.corr().abs(), annot=True, square=True, annot_kws={'size': 7}, fmt='.2f',cbar_kws={"shrink": 0.73}) 
axs.tick_params(axis='x', labelsize=8)  
axs.tick_params(axis='y', labelsize=8)  
axs.set_title('Correlation Chart', size=10)
plt.tight_layout()  
plt.show()
```
